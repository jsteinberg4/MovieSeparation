```python
import pandas as pd
```

```python
# Columns to drop, decided in IMDB_Movies.ipynb
score_cols = ["avg_vote", "votes", "metascore", "reviews_from_critics", "reviews_from_users"]
money_cols = ["budget", "usa_gross_income", "worlwide_gross_income"]
description_col = ["description", "duration", "language", "genre", "production_company", "country"]

movies = (pd.read_csv('../Datasets/IMDb movies.csv')
          .drop(columns=[*score_cols, *money_cols, *description_col]))
```

```python
test_movies = movies[:1].rename({'imdb_title_id': 'imdb_id'}, axis=1)
test_movies
```

```python
from Notebooks.movie import Movie
from Notebooks.person import Person
movie_cols = ['imdb_id', 'title', 'original_title', 'year', 'date_published']

row = test_movies.iloc[0]
miss_jerry = Movie(**row[movie_cols])
director = Person(row['director'])
writer = Person(row['writer'])
actors = [Person(name.strip()) for name in row['actors'].split(',')]
```

```python
print(miss_jerry, director, writer, actors, sep='\n')
```
